```
import pandas as pd
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```
#--------Define_tensors----------------------
tensor_one = torch.tensor([1,2,3,4,5])
tensor_two = torch.tensor([2,3,4,5,6])
```

```
#--------------Add_two_tensors-----------------
addition = tensor_one + tensor_two
print("The Addition is:{} ".format(addition))
```

```
#-----------Multiply_tensors------------
multiplication  = tensor_one * tensor_two
print("The Multiplication is:{} ".format(multiplication))
```

```
#----------Substract_tensors---------------
substraction = tensor_one - tensor_two
print("The Substraction is:{} ".format(substraction))
```

```
#----------Divide_tensors-----------------
division = tensor_one // tensor_two
print("The division is:{} ".format(division))
```

```
#-----------Multiply_tensor_by_scaler---------------
print("The tensor one is:{}".format(tensor_one))
scalar_mul  = tensor_one*5
print("The multiplication by scaler is:{}".format(scalar_mul))
```

```
#-----------Add_scaler_value_to_tensor-------------------
add_scalar = tensor_one + 5
print("The tensor_after adding scalar 5 is:{}".format(add_scalar))
```

```
#----------Substarct_scalar_value_from_tensor-----------------
print("The tensor2 is:{}".format(tensor_two))
sub_scalar  = tensor_two - 1
print("The tesnor after substracting value from it is:{}".format(sub_scalar))
```

```
#---------------Calculate_dot_product_of_two_tensors------------------
dot_product = torch.dot(tensor_one,tensor_two)
print("The dot product of two tensors is:{}".format(dot_product))
```

```
#--------------Generate_data_points_within_certain_range_using_pytorch------------------
numbers = torch.linspace(1,10,100)
print("The tensor of numbers are \n:{}".format(numbers))
```

The linspace function generate certain numbers within certain specified range. Here in above code snippet i have ganerated 100 numbers between interval (1-10).

```
#---------------Plot_generated_numbers--------------------
x_coordinates = torch.linspace(1,10,100)
y_coordinates = x_coordinates+5
plt.figure(figsize=(10,5))
plt.plot(x_coordinates.numpy(),y_coordinates.numpy())
plt.xlabel('x_coordinates')
plt.ylabel('y_coordinates')
```

```
#-------------------------Use_of_exponential_function-----------------------
plt.figure(figsize=(10,5))
exponential_coordinates = torch.exp(x_coordinates)
plt.plot(x_coordinates.numpy(),exponential_coordinates.numpy())
```

```
#------------------------Use_of_sin_function----------------
plt.figure(figsize=(10,5))
sin_coordinates = torch.sin(x_coordinates)
plt.plot(x_coordinates.numpy(),sin_coordinates.numpy())
```

```
#------------------------Use_of_cos_function----------------
plt.figure(figsize=(10,5))
cos_coordinates = torch.cos(x_coordinates)
plt.plot(x_coordinates.numpy(),cos_coordinates.numpy())
```

```
#------------------------Use_of_cos_function----------------
plt.figure(figsize=(10,5))
tan_coordinates = torch.tan(x_coordinates)
plt.plot(x_coordinates.numpy(),tan_coordinates.numpy())
```
